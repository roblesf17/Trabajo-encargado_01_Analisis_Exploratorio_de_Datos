1.- Carga el conjunto de datos

```r
recaudacion <- read.csv("IngresosRecaudadosSUNAT.csv", sep = ";", header = TRUE, fileEncoding="utf-8")
```

2.- Revisar los seis primeros datos por defecto 

```r
head(recaudacion)
```

3.- Verificar la estructura de los datos

```r
str(recaudacion)
```

4.- Revisar un resumen de los datos

```r
summary(recaudacion)
```

5.- Revisar un resumen de los datos en columnas

```r
library(mlr)
summarizeColumns(recaudacion)
```

6.- Revisar graficamente el porcentaje de nulos

```r
library(VIM)
library(colorspace)
library(grid)
library(data.table)
aggr(recaudacion,numbers=TRUE, plot = T)
```

7.- Visualizar la tabla y frecuencia con formato SAS

```r
library(gmodels)
library(gdata)
CrossTable(recaudacion$Codigo_Fuente, format="SAS")
```

8.- Visualizar la tabla y frecuencia con formato SPSS

```r
CrossTable(recaudacion$Codigo_Fuente, format="SPSS")
```

9.- Visualizar gráficos de resumen de las variables: gráfico de barras

```r
frec <- table(recaudacion$Codigo_Fuente)
barplot(frec, main="INGRESOS RECAUDADOS POR LA SUNAT - 2019 \n SEGÚN CANTIDAD DE TIPOS DE FUENTES DE INGRESOS", xlab="Codigo_Fuente", ylab="Total",col = c("lightblue", 
        "lightcyan"),legend = rownames(frec),ylim=c(0,15))
```

10.- Visualizar gráficos de resumen de las variables: pie

```r
frec <- table(recaudacion$Fuente_Ingresos)
piepercent <- round(100*frec/sum(frec), 2)
pie(frec,piepercent,main="Porcentaje de los Ingresos Recaudados por la SUNAT - 2019 \n SEGÚN CANTIDAD DE TIPOS DE FUENTES DE INGRESOS", xlab="Fuente_Ingresos",col = rainbow(length(frec)))
   legend("topright",  c("Aduaneros", "Contribuciones Sociales", "No Tributarios", "Internos"), cex = 0.7, fill = rainbow(length(frec)))
```

11.- Visualizar gráficos de resumen de las variables: pie 3D

```r
library(plotrix)
lbls1 <- paste(names(table(recaudacion$Fuente_Ingresos)), "\n", round((prop.table(table(recaudacion$Fuente_Ingresos))*100),2),"%", sep="")
pie3D(table(recaudacion$Fuente_Ingresos), labels = lbls1,explode=0.15, main="Porcentaje de los Ingresos Recaudados por la SUNAT - 2019 \n según los tipos de Funte")
```

12.- Visualizar gráficos de resumen de las variables: grafico de puntos

```r
dotchart(table(recaudacion$Codigo_Fuente), cex=0.7, main="Porcentaje de los Ingresos Recaudados por la SUNAT - 2019 ", xlab="Fuente_Ingresos")
```

13.- Generar tablas dinamicas o de contingencia

```r
Tributos <- recaudacion$Concepto
Fuentes <- recaudacion$Codigo_Fuente
CrossTable(Tributos,Fuentes,format="SAS")
```

```r
Tributos <- recaudacion$Concepto
Fuentes <- recaudacion$Codigo_Fuente
CrossTable(Tributos,Fuentes,format="SPSS")
```

15.- Representación de variables cuantitativas: 

```r
frec <- table(recaudacion$Fuente_Ingresos)
frec
legendPlot <- recaudacion$Total
bp1  <- barplot(
    frec,
    main="Ingresos Recaudados por la SUNAT \n según Tipos de Fuentes de Ingresos- 2019",
    xlab="",
    ylab="Total",
    col = rainbow(length(legendPlot)),
    ylim=c(0,15),
    las=2,
    cex.names=0.6
)
```

16.- Visualizar linea de distribucion de los conceptos de tributos por el tipo de fuente

```r
Tributos <- recaudacion$Concepto
Fuentes <- recaudacion$Codigo_Fuente
frec <- table(Tributos,Fuentes)
frec
legendPlot=levels(Tributos)
bp1  <- barplot(frec,beside = T, main="CANTIDAD DE TIPOS DE FUENTES DE ", xlab="TIPO DE INGRESOS", ylab="Cantidad"
               ,col = rainbow(length(legendPlot)), ylim = c(0,10) )
legend("topright",  legend=legendPlot, cex = 0.8,fill = rainbow(length(legendPlot)))
```
